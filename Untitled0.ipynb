**Usar libreria pandas para analisis**

```python
import pandas as pd
# carga de datos
df = pd.read_csv('ventas.csv')
```

```python
#ver primeras filas
print(df.head())
```

```python
#ver primeras filas
print(df.info())
```

```python
#Procesamiento (limpieza y transformacion)
import pandas as pd
df["fecha"] = pd.to_datetime(df["fecha"])

#crear columna de ingreso
df["ingreso"] = df["cantidad"] * df["precio"]

print(df)
```

```python
#Analisis de datos
#Total de ingresos

total_ingreso = df["ingreso"].sum()
print("Total de ingresos: ", total_ingreso)
```

```python
#Producto mas vendido
ventas_producto = df.groupby("producto")["cantidad"].sum()
print(ventas_producto)
```

```python
# ventas por dia
ventas_dia = df.groupby("fecha")["ingreso"].sum()
print(ventas_dia)

```

```python
#visualizacion
import matplotlib.pyplot as plt

ventas_producto.plot(kind="bar", title="Ventas por producto")
plt.show()
```
